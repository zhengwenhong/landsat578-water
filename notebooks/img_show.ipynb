```python
import config
import sys
sys.path.append(config.root)
import matplotlib.pyplot as plt
import numpy as np
from utils.geotif_io import readTiff, writeTiff
from utils.imgShow import imgShow
```

```python
# sen2 images 
scene_num = 'l5_scene_02'
path_img = config.root + '/data/dataset-l578/scene/' + scene_num+ '.tif'
path_truth = config.root + '/data/dataset-l578/truth/' + scene_num+ '_truth.tif'

image, image_info = readTiff(path_in=path_img)
truth, truth_info = readTiff(path_in=path_truth)
image = np.float32(np.clip(image/10000, a_min=0, a_max=1)) 
print(image.shape)
print(truth.shape)
```

```python
plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
imgShow(image, extent=image_info['geoextent'], color_bands=(2,1,0), clip_percent=2)
plt.subplot(1,2,2)
imgShow(truth, color_bands=(0,0,0), clip_percent=2)
```

### Save to .png

```python
## --- save scene ---
fig = plt.figure(figsize=(4,4),dpi=100)
imgShow(image, color_bands=(2,1,0), clip_percent=2)
plt.axis('off')
path_fig = config.root + '/figures/dataset/'+ scene_num + '.png'
fig.savefig(path_fig)
## --- save truth ---
fig = plt.figure(figsize=(4,4),dpi=100)
imgShow(truth, color_bands=(0,0,0), clip_percent=2)
plt.axis('off')
path_fig = config.root + '/figures/dataset/'+ scene_num + '_truth.png'
fig.savefig(path_fig)

```
